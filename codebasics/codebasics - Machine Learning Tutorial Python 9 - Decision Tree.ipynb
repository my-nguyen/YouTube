## Part 1: use LabelEncoder

```python
import pandas as pd

df = pd.read_csv('salaries9.csv')
df
```

```python
df.company.unique()
```

```python
df.job.unique()
```

```python
df.degree.unique()
```

```python
X = df.drop('salary_more_then_100k', axis='columns')
y = df.salary_more_then_100k
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
X['company_n'] = label.fit_transform(df.company)
X['job_n'] = label.fit_transform(df.job)
X['degree_n'] = label.fit_transform(df.degree)
X
```

```python
X = X.drop(['company', 'job', 'degree'], axis='columns')
X
```

# for company
* 2 = google
* 0 = abc pharma
* 1 = facebook

# for job
* 2 = sales executive
* 0 = business manager
* 1 = computer programmer

# for degree
* 0 = bachelors
* 1 = masters

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
# predict salary for a person working for facebook as a business manager with a masters degree
model.predict([[1, 0, 1]])
```

result 1 means this person likely makes salary more than 100k

```python
y_test
```

## Part 2: use OneHotEncoder instead

```python
import pandas as pd

df = pd.read_csv('salaries9.csv')
df
```

```python
X = df.drop('salary_more_then_100k', axis='columns')
y = df.salary_more_then_100k
X
```

```python
# Encode categories using sklearn.preprocessing LabelEncoder, OneHotEncoder and ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([('my_ct', OneHotEncoder(), [0, 1, 2])], remainder='passthrough')
X = ct.fit_transform(X)
X
```

# for company
* [0, 0, 1] = google
* [1, 0, 0] = abc pharma
* [0, 1, 0] = facebook

# for job
* [0, 0, 1] = sales executive
* [1, 0, 0] = business manager
* [0, 1, 0] = computer programmer

# for degree
* [1, 0] = bachelors
* [0, 1] = masters

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
# likelihood of making more than 100k of a person working for google as a computer programmer with a bachelors degree
model.predict([[0, 0, 1, 0, 1, 0, 1, 0]])
```

```python
# likelihood of making more than 100k of a person working for facebook as a business manager with a bachelors degree
model.predict([[0, 1, 0, 1, 0, 0, 1, 0]])
```

## Part 3: Exercise - Build decision tree model to predict survival based on certain parameters

CSV file is available to download at https://github.com/codebasics/py/blob/master/ML/9_decision_tree/Exercise/titanic.csv

##### In this file using following columns build a model to predict if person would survive or not,

1. Pclass
1. Sex
1. Age
1. Fare

##### Calculate score of your model

```python
import pandas as pd

df = pd.read_csv('titanic9.csv')
df
```

```python
# retain only relevant columns
df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
df
```

```python
y = df.Survived
X = df.drop('Survived', axis='columns')
X
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
X['Sex_n'] = label.fit_transform(df.Sex)
X
```

```python
X = X.drop('Sex', axis='columns')
X
```

```python
median_age = round(df.Age.mean())
median_age
```

```python
X.Age = X.Age.fillna(median_age)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
